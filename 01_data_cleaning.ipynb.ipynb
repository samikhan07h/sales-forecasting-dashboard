```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw_sales.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df = df.sort_values('Date')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Sales'] = df['Sales'].ffill()
df.head()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
daily_sales = df.groupby('Date')['Sales'].sum().reset_index()
daily_sales.head()
```

```python
daily_sales.to_csv("../data/cleaned_sales.csv", index=False)
```
